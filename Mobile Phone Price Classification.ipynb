```python
data = data.drop(['m_dep','weight'],axis=1)
```

```python
data.head()
```

```python
data = data[[ 'touch_screen','three_g','four_g', 'dual_sim','wifi','bluetooth','ram','primary_cam','front_cam','talk_time','battery_power','clock_speed',
        'n_cores',
       'px_height', 'px_width', 'sc_h', 'sc_w', 
        'price_range']]
```

```python
data.head()
```

```python
X  = data.drop('price_range',axis=1)
y = data['price_range']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
pickle.dump(sc,(open('sc.pkl','wb')))
X.iloc[:,6:] = sc.fit_transform(X.iloc[:,6:])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train,y_train)
```

```python
ypred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,ypred))
print(confusion_matrix(y_test,ypred))
print(accuracy_score(y_test,ypred))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train,y_train)
ypred1 = classifier.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,ypred1))
print(confusion_matrix(y_test,ypred1))
print(accuracy_score(y_test,ypred1))
```

```python
from sklearn.svm import SVC
SVC = SVC()
SVC.fit(X_train,y_train)
y_pred1 = SVC.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(accuracy_score(y_test,y_pred1))
```
